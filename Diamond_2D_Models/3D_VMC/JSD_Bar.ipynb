```python
import matplotlib.pyplot as plt

# Data
Model = ["Our Model", "SCUNET (Pre)", "SCUNET (Train)", "Nature", "BM3D"]
JSD = [0.00025433622, 0.000138, 0.000515, 0.000504, 0.000316]

# Create bar plot
plt.figure(figsize=(8, 5))
bars = plt.bar(Model, JSD, color='skyblue', edgecolor='black')

# Add value labels on top of bars
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height,
             f'{height:.2e}', ha='center', va='bottom', fontsize=9)

# Axis labels and title
plt.ylabel("JSD")
plt.title("JSD Comparison Across Models")
plt.xticks(rotation=15)
plt.tight_layout()

plt.show()
```
